```python
import pandas as pd
import numpy as np
```

```python
analytics = pd.read_csv(
    '../data/cleaned/customer_loan_analytics.csv'
)

analytics.head()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

CREATE EXECUTIVE KPIs

```python
#KPI 1 — TOTAL LOAN PORTFOLIO
total_loan_portfolio = analytics[
    'loan_amount'
].sum()

total_loan_portfolio
```

```python
#KPI 2 — TOTAL CUSTOMERS
total_customers = analytics[
    'customer_id'
].nunique()
total_customers
```

```python
#KPI 3 — DEFAULT RATE
default_rate = (
    analytics['high_risk_flag']
    .mean()
) * 100

default_rate
```

```python
#KPI 4 — MISSED PAYMENT RATE
missed_payment_rate = (
    analytics['missed_payment_flag']
    .mean()
) * 100

missed_payment_rate
```

```python
#KPI 5 — AVG LOAN AMOUNT
avg_loan_amount = analytics[
    'loan_amount'
].mean()

avg_loan_amount
```

```python
#KPI TABLE
dashboard_kpis = pd.DataFrame({

    'metric': [
        'Total Loan Portfolio',
        'Total Customers',
        'Default Rate',
        'Missed Payment Rate',
        'Average Loan Amount'
    ],

    'value': [
        total_loan_portfolio,
        total_customers,
        default_rate,
        missed_payment_rate,
        avg_loan_amount
    ]
})

dashboard_kpis
```

## Executive Dashboard Insight

The dashboard KPIs provide a high-level overview of portfolio health and lending risk exposure.

CREATE DASHBOARD CHART DATASETS

```python
#LOAN STATUS
loan_status_summary = (
    analytics['loan_status']
    .value_counts()
    .reset_index()
)

loan_status_summary.columns = [
    'loan_status',
    'count'
]

loan_status_summary
```

```python
loan_status_summary.to_csv(
    '../data/cleaned/loan_status_summary.csv',
    index=False
)
```

```python
#INCOME SEGMENT ANALYSIS
income_segment_summary = (
    analytics.groupby(
        'income_segment'
    )['loan_amount']
    .mean()
    .reset_index()
)

income_segment_summary
```

## Customer Segmentation Insight

Middle-income and salaried customers appear to dominate the lending portfolio, indicating lower-risk lending behavior.

```python
income_segment_summary.to_csv(
    '../data/cleaned/income_segment_summary.csv',
    index=False
)
```

```python
#EMPLOYMENT ANALYSIS
employment_summary = (
    analytics.groupby(
        'employment_type'
    )['loan_amount']
    .mean()
    .reset_index()
)

employment_summary
```

```python
employment_summary.to_csv(
    '../data/cleaned/employment_summary.csv',
    index=False
)
```
