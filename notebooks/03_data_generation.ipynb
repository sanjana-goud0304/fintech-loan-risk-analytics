```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker()

print(fake.name())
print(fake.city())
```

```python
num_customers = 10000

customers = []

employment_types = [
    'Salaried',
    'Self-Employed',
    'Bussiness',
    'Freelancer'
]

for customer_id in range(1, num_customers + 1):
    
    customers.append({
        'customer_id': customer_id,
        'first_name' : fake.first_name(),
        'last_name' : fake.last_name(),
        'gender' : random.choice(['Male', 'Female']),
        'date_of_birth' : fake.date_of_birth(minimum_age=21, maximum_age=60),
        'city' : fake.city(),
        'state' : fake.state(),
        'employment_type' : random.choice(employment_types),
        'monthly_income' : random.randint(25000, 200000),
        'signup_date' : fake.date_between(start_date='-3y', end_date='today')

})
    
customers_df = pd.DataFrame(customers)

customers_df.head()
```

```python
customers_df.to_csv(
    "../data/raw/customers.csv",
    index=False
)
```

```python
num_loans = 15000

loan_statuses = [
    "Active",
    "Closed",
    "Default"
]

loans = []

for loan_id in range(1, num_loans + 1):

    customer_id = random.randint(1, num_customers)

    loans.append({
        "loan_id": loan_id,
        "customer_id": customer_id,
        "loan_amount": random.randint(50000, 1000000),
        "interest_rate": round(random.uniform(8, 24), 2),
        "tenure_months": random.choice([6, 12, 24, 36, 48]),
        "loan_status": random.choice(loan_statuses),
        "disbursement_date": fake.date_between(start_date='-2y', end_date='today')
    })

loans_df = pd.DataFrame(loans)

loans_df.head()
```

```python
print(len(loans_df))
```

```python
loans_df.to_csv(
    '../data/raw/loans.csv',
    index=False
)
```

```python
num_transactions = 50000

transaction_types = [
    "UPI",
    "Card",
    "Bank Transfer",
    "Wallet"
]

merchant_categories = [
    "Food",
    "Shopping",
    "Travel",
    "Bills",
    "Entertainment",
    "Healthcare"
]

transactions = []

for transaction_id in range(1, num_transactions + 1):

    transactions.append({
        "transaction_id": transaction_id,
        "customer_id": random.randint(1, num_customers),
        "transaction_date": fake.date_between(start_date='-1y', end_date='today'),
        "transaction_type": random.choice(transaction_types),
        "merchant_category": random.choice(merchant_categories),
        "amount": round(random.uniform(100, 50000), 2)
    })

transactions_df = pd.DataFrame(transactions)

transactions_df.head()
```

```python
transactions_df.to_csv(
    '../data/raw/transactions.csv',
    index=False
)
```

```python
repayments = []

repayment_statuses = [
    "On-Time",
    "Late",
    "Missed"
]

for repayment_id in range(1, 20001):

    repayments.append({
        "repayment_id": repayment_id,
        "loan_id": random.randint(1, num_loans),
        "due_date": fake.date_between(start_date='-1y', end_date='today'),
        "payment_date": fake.date_between(start_date='-1y', end_date='today'),
        "amount_paid": round(random.uniform(1000, 50000), 2),
        "repayment_status": random.choice(repayment_statuses)
    })

repayments_df = pd.DataFrame(repayments)

repayments_df.head()
```

```python
repayments_df.to_csv(
    "../data/raw/repayments.csv",
    index=False
)
```
