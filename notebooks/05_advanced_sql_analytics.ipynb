# Day 5 — Advanced SQL Analytics

## Objective
Perform advanced fintech analytics using SQL.

## Concepts Covered
- CTEs
- Window Functions
- KPIs
- Risk Analysis
- Customer Segmentation
- Trend Analysis

## Business Insight

The default rate KPI helps identify overall portfolio risk. A higher default rate may indicate weak credit approval policies.

## Customer Segmentation Insight

High-income customers tend to receive larger loan amounts, indicating higher lending confidence from the institution.

## Fraud Analysis Insight

Customers with unusually large transaction volumes may require additional monitoring for suspicious financial activity.
